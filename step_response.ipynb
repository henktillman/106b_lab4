```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.concatenate((np.zeros(10), np.ones(10)))
c = 30.0 * (x / 90) + 25
t = np.arange(20)

plt.close('all')

f, (a1, a2) = plt.subplots(2, sharex=True)
a1.set_title('Step response of flex controller')
a1.plot(t, x)
a1.set_ylabel('Error')
a2.plot(t, c)
a2.set_ylabel('Control')
a2.set_xlabel('Timestep')

plt.show()
```

```python
x = np.concatenate((np.zeros(10), np.ones(10)))
exp_avg_x = [0]
for i in range(1, 20):
    exp_avg_x.append(0.8*exp_avg_x[i-1] + 0.2*x[i])
exp_avg_x = np.array(exp_avg_x)
c = 70 + 6*exp_avg_x
t = np.arange(20)

plt.close('all')

f, (a1, a2) = plt.subplots(2, sharex=True)
a1.set_title('Step response of pressure controller')
a1.plot(t, x)
a1.set_ylabel('Error')
a2.plot(t, c)
a2.set_ylabel('Control')
a2.set_xlabel('Timestep')

plt.show()
```
